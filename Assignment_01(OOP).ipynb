# Step 2: Implement Inheritance (Student & Instructor)

```python
class User:
    def __init__(self, name, email):
        self.name = name
        self._email = email

    def display_profile(self):
        print(f"Name: {self.name}")
        print(f"Email: {self._email}")


# Student class
class Student(User):
    def __init__(self, name, email):
        super().__init__(name, email)
        self.__enrolled_courses = []

    def enroll(self, course_name):
        self.__enrolled_courses.append(course_name)
        print(f"{self.name} has been enrolled in '{course_name}'.")

    def view_courses(self):
        if self.__enrolled_courses:
            print("Enrolled Courses:")
            for course in self.__enrolled_courses:
                print(f"- {course}")
        else:
            print("No courses enrolled yet.")

    # Method overriding (Polymorphism)
    def display_profile(self):
        print("=== Student Profile ===")
        print(f"Name: {self.name}")
        print(f"Email: {self._email}")
        print("Badge: Student 🎓")


# Instructor class
class Instructor(User):
    def __init__(self, name, email, assigned_department):
        super().__init__(name, email)
        self.assigned_department = assigned_department

    # Method overriding
    def display_profile(self):
        print("=== Instructor Profile ===")
        print(f"Name: {self.name}")
        print(f"Email: {self._email}")
        print(f"Department: {self.assigned_department}")
```

```python
student1 = Student("Alice", "alice@example.com")
student1.display_profile()
student1.enroll("Python Programming")
student1.enroll("Data Science")
student1.view_courses()

print()

instructor1 = Instructor("Bob", "bob@example.com", "Software Engineering")
instructor1.display_profile()
```

# Step 3: Create the Course Class

```python
class Course:
    def __init__(self, course_name, duration, instructor):
        self.course_name = course_name
        self.duration = duration
        self.instructor = instructor   # expects an Instructor object

    def course_details(self):
        print("=== Course Details ===")
        print(f"Course Name: {self.course_name}")
        print(f"Duration: {self.duration}")
        print(f"Instructor: {self.instructor.name}")
```

```python
instructor1 = Instructor("Bob", "bob@example.com", "Data Science")

course1 = Course("Machine Learning", "6 Days", instructor1)
course1.course_details()
```

# Step 4: Bring it All Together (The Execution)

```python
# 1. Instantiate an Instructor
instructor1 = Instructor("Dr. Smith", "smith@example.com", "Data Science")

# 2. Instantiate a Course with the instructor
course1 = Course("Machine Learning", "6 Days", instructor1)

# 3. Instantiate two Student objects
student1 = Student("Alice", "alice@example.com")
student2 = Student("John", "john@example.com")

# 4. Enroll both students in the course
student1.enroll(course1.course_name)
student2.enroll(course1.course_name)

print("\n")

# 5. Call display_profile() on all users (Polymorphism)
instructor1.display_profile()
print()
student1.display_profile()
print()
student2.display_profile()

print("\n")

# Optional: View enrolled courses
student1.view_courses()
print()
student2.view_courses()

print("\n")

# 6. Call course_details()
course1.course_details()
```

# What this demonstrates:
1. Encapsulation → private __enrolled_courses
2. Inheritance → Student and Instructor from User
3. Polymorphism → overridden display_profile() methods
4. Object interaction → Course uses an Instructor object

```python
class Student(User):
    def __init__(self, name, email):
        super().__init__(name, email)
        self.__enrolled_courses = []   # private attribute

    def enroll(self, course_name):
        self.__enrolled_courses.append(course_name)
        print(f"{self.name} has been enrolled in '{course_name}'.")

    def unenroll(self, course_name):
        if course_name in self.__enrolled_courses:
            self.__enrolled_courses.remove(course_name)
            print(f"{self.name} has been unenrolled from '{course_name}'.")
        else:
            print(f"Error: {self.name} is not enrolled in '{course_name}'.")

    def view_courses(self):
        if self.__enrolled_courses:
            print("Enrolled Courses:")
            for course in self.__enrolled_courses:
                print(f"- {course}")
        else:
            print("No courses enrolled yet.")

    def display_profile(self):
        print("=== Student Profile ===")
        print(f"Name: {self.name}")
        print(f"Email: {self._email}")
        print("Badge: Student 🎓")
```

```python
student1 = Student("Alice", "alice@example.com")

student1.enroll("Python Programming")
student1.view_courses()

student1.unenroll("Python Programming")   # valid removal
student1.unenroll("Data Science")         # not enrolled → error
```
